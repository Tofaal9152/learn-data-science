```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave Example')
plt.legend()
plt.show()
```
